```python
# fibonacci numbers in the paradigm of 

def Fibonacci(n): 
    if n < 0: 
        print("Incorrect input") 
    # First Fibonacci number is 0 
    elif n==1: 
        return 0
    # Second Fibonacci number is 1 
    elif n==2: 
        return 1
    else: 
        return Fibonacci(n-1)+Fibonacci(n-2) 
```

```python
start = time.time()
print(Fibonacci(40))
end = time.time()
print(timedelta(seconds = end - start))
```

```python
# fibonacci numbers in the pradigm of dynamic programming

FibArray = [0,1] 
  
def fibonacci(n): 
    if n < 0: 
        print("Incorrect input") 
    elif n <= len(FibArray): 
        return FibArray[n-1] 
    else: 
        temp_fib = fibonacci(n-1)+fibonacci(n-2) 
        FibArray.append(temp_fib) 
        return temp_fib 
```

```python
start = time.time()
print(fibonacci(40))
end = time.time()
print(timedelta(seconds = end - start))
```

## Main libraries used in data analytics in Python

```python
# numerical python
import numpy as np
# pandas dataframe - main tool of data scientist (mainstream ds)
import pandas as pd

# package for plotting, one may also consider plotly - more complicated, but more fancy
import matplotlib.pyplot as plt
```

```python
# create sequence of integers from 1 to n
t = np.arange(1, 50)
print(t)
```

```python
# create random noise
epsilon = np.random.normal(0, 1, len(t))
# create variable Y
Y = 10*np.ones(len(t)) + 0.5*t + epsilon
print(Y)
```

```python
# create pd.DataFrame
df = pd.DataFrame(
    {
        "Y": Y,
        "intercept": np.ones(len(t)),
        "t": t,
        "eps": epsilon
    }
)

display(df.head())
```

<h1><center>
Let's estimate OLS linear trend model
<br/>
$\hat{\beta} = (X^T X)^{-1} X^T Y$
</center><h1>



```python
X = df[["intercept", "t"]]
display(X)
print(X.shape)
print(X.transpose().shape)
```

```python
XTX = np.dot(X.values.transpose(), X.values)
print(XTX)
```

```python
beta_hat = np.dot(np.linalg.inv(XTX), X.values.transpose())
print("(XTX)^(-1)*XT: ", beta_hat)
print("(XTX)^(-1)*XT: ", beta_hat.shape)
beta_hat = np.dot(beta_hat, Y)
print("(XTX)^(-1)*XT*Y: ", beta_hat)
print("beta_1: ", beta_hat[0])
print("beta_2: ", beta_hat[1])
```

```python
# make forecasts of Y
df["Y_pred"] = beta_hat[0] + beta_hat[1]*df["t"]
```

```python
plt.figure(figsize = (20, 10))
plt.plot(df["t"], df["Y"], label = "Y actual")
plt.plot(df["t"], df["Y_pred"], label = "Y predicted")
plt.xlabel("time")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```

```python
class Animal:
    
    # constructor
    def __init__(self, color):
        self.color = color
        
    def make_noise(self):
        print("Some noise of the animal")

# Inheritance
class Dog(Animal):
    
    def __init__(self, color, height, breed):
        super(Dog, self).__init__(color)
        self.height = height
        self.breed = breed
        
    def make_noise(self):
        print("Bark Bark")
```

```python
some_animal = Animal(color = "black")
some_animal.make_noise()
print(some_animal.color)
```

```python
pomeranian = Dog(color = "white", height = "20cm", breed = "pomeranian")
pomeranian.make_noise()
print(pomeranian.breed)
print(pomeranian.color)
print(pomeranian.height)
```

# Monte Carlo simulations

$ 
\int_{0}^{1} e^{-x} dx = \frac{e - 1}{e} = 0.63212
$ 

```python
# create f(x) = e^x
x = np.linspace(0, 1, 100)
y = np.exp(-x)
```

```python
plt.figure(figsize = (20, 10))
plt.plot(x, y)
plt.ylim(0, 1.75)
plt.grid()
plt.show()
```

```python
counter = 0
num_sim = 0

for i in range(len(x)):
    y_ax_uniform = np.random.uniform(0, np.max(y), 1000)    
    counter += len(np.where(y_ax_uniform <= y[i])[0])
    num_sim += len(y_ax_uniform)

print(num_sim)
print(counter)
print(counter/num_sim)
```

## Random walk simulation

```python
length = 300

def generate_rw(length):
    
    temp_eps = np.random.normal(0, 1, length)
    x_t = [0]
    for i in range(1, length):
        x_t.append(x_t[i-1] + temp_eps[i])
    return np.array(x_t)

def generate_n_path(n_path, length):
    
    temp_dict = dict.fromkeys(np.arange(n_path))
    
    for key in temp_dict.keys():
        temp_list = generate_rw(length)
        temp_dict[key] = temp_list
    
    temp_df = pd.DataFrame(temp_dict, np.arange(length))
    return temp_df
```

```python
random_walk_pathes = generate_n_path(1000, length)
```

```python
plt.figure(figsize = (20, 10))
for colname in random_walk_pathes.columns:
    plt.plot(random_walk_pathes[colname])
plt.grid()
plt.show()
```

# Newbold-Granger experiment

```python
import statsmodels.api as sm
```

```python
length = 100
def generate_yx_data(intercept, slope, length):
    
    temp_u = np.random.normal(0, 1, length)
    
    y_t = [0]
    for i in range(1, length):
        y_t.append(intercept + slope*y_t[i-1] + temp_u[i])
    temp_df = pd.DataFrame(
        {"y_t": y_t}
    )
    temp_df = sm.add_constant(temp_df)
    return temp_df
```

```python
def estimate_ols(y_int, y_slope, x_int, x_slope, length):
    df_y = generate_yx_data(y_int, y_slope, length)
    df_x = generate_yx_data(x_int, x_slope, length)
    
    df_ols = pd.DataFrame(
        {
            "const": df_y["const"],
            "y_t": df_y["y_t"],
            "x_t": df_x["y_t"]
        }
    )
    
    model = sm.OLS(df_ols["y_t"],df_ols[["const", "x_t"]])
    results = model.fit()
    df_ols["y_t_pred"] = model.predict(results.params)
    
    return results
```

```python
slope_list = []
slope_pval_list = []

num_exper = 100

for i in range(num_exper):
    
    temp_ols = estimate_ols(0.5, 0.9, 0.3, 0.9, length)
    slope_list.append(temp_ols.params["x_t"])
    slope_pval_list.append(temp_ols.t_test("x_t").pvalue)
    
df_test = pd.DataFrame(
    {
        "slope": slope_list,
        "pval": slope_pval_list
    }
)
df_test["pval"] = df_test["pval"].apply(lambda x: float(x))
```

```python
plt.figure(figsize = (20, 10))
df_test["slope"].plot("hist", bins = 10)
plt.grid()
plt.show()
```
